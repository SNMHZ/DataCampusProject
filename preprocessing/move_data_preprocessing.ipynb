```python
import os

print()

path = os.getcwd() + "/common_file/move_file"

file_list = os.listdir(path)
file_list_py = [file for file in file_list]

data_file = []
path_s_list = [] #세부 경로

print(file_list_py)

for i in range(len(file_list_py)) :
  path_s = path + '/' + file_list_py[i]
  path_s_list.append(path_s)


for i in range(len(file_list_py)) :
  path_s = path + '/' + file_list_py[i]
  file_list_s = os.listdir(path_s)
  file_list_py_s = [path_s + '/' + file  for file in file_list_s]
  
  for i in range(len(file_list_py_s)) :
    print(file_list_py_s[i])

  
  data_file.append(file_list_py_s)  


```

```python
def chain(*iterables):
    # chain('ABC', 'DEF') --> ['A', 'B', 'C', 'D', 'E', 'F']
    for it in iterables:
        for element in it:
            yield element
```

```python
import numpy as np
import itertools

result_list = list(itertools.chain(*data_file))
result_list.reverse()
result_list
```

```python
import pandas as pd

#빈 데이터 프레임 만들기
df_result6 = pd.DataFrame({'시간대구분': [6],
                              '집계구코드': [1101072010001],
                              '총생활인구수': [0]})
df_result9 = pd.DataFrame({'시간대구분': [9],
                              '집계구코드': [1101072010001],
                              '총생활인구수': [0]})
df_result12 = pd.DataFrame({'시간대구분': [12],
                              '집계구코드': [1101072010001],
                              '총생활인구수': [0]})
df_result15 = pd.DataFrame({'시간대구분': [15],
                              '집계구코드': [1101072010001],
                              '총생활인구수': [0]})
df_result18 = pd.DataFrame({'시간대구분': [18],
                              '집계구코드': [1101072010001],
                              '총생활인구수': [0]})
df_result21 = pd.DataFrame({'시간대구분': [21],
                              '집계구코드': [1101072010001],
                              '총생활인구수': [0]})
```

```python
for i in range(len(result_list)):

    df = pd.read_csv(result_list[i])
    df = df[['시간대구분', '집계구코드', '총생활인구수']]

    df_time6 = df['시간대구분'] == 6
    df_time9 = df['시간대구분'] == 9
    df_time12 = df['시간대구분'] == 12
    df_time15 = df['시간대구분'] == 15
    df_time18 = df['시간대구분'] == 18
    df_time21 = df['시간대구분'] == 21

    df_time_s6 = df[df_time6]
    df_time_s9 = df[df_time9]
    df_time_s12 = df[df_time12]
    df_time_s15 = df[df_time15]
    df_time_s18 = df[df_time18]
    df_time_s21 = df[df_time21]

    df_result6 = pd.concat([df_result6, df_time_s6])
    df_result9 = pd.concat([df_result9, df_time_s9])
    df_result12 = pd.concat([df_result12, df_time_s12])
    df_result15 = pd.concat([df_result15, df_time_s15])
    df_result18 = pd.concat([df_result18, df_time_s18])
    df_result21 = pd.concat([df_result21, df_time_s21])

    print(len(df_result6), len(df_result9), len(df_result12), len(df_result15), len(df_result18), len(df_result21))

    print("====================================")
```

```python
result = pd.concat([df_result6,df_result9])
result = pd.concat([result,df_result12])
result = pd.concat([result,df_result15])
result = pd.concat([result,df_result18])
result = pd.concat([result,df_result21])
result = result[['집계구코드', '총생활인구수']]
result
```

```python
grouped = result.groupby('집계구코드').mean()
grouped
```

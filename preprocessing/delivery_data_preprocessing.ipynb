```python
import os

col1 = []
col2 = []
col3 = []
col4 = []



file = os.getcwd() + "\\common_file\\11_2019년_성연령별인구.txt"
f = open(file, 'r')
line = f.readline()

while True:

    line = f.readline()
    if not line: 
      break
    l = line.split('^')

    col1.append(l[0])
    col2.append(l[1])
    col3.append(l[2])
    col4.append(l[3])

f.close()
```

```python
for i in range(len(col4)) :
  if col4[i] == 'N/A\n' or col4[i] == 'N/A' :
      col4[i] = '2.5'
  
  col4[i] = float(col4[i])
```

```python
x_list = []
y_list = []
for geo in family1['geometry']:
  x_list.append( int(round(geo.x)) )
  y_list.append( int(round(geo.y)) )

family1['utmk_x'] = x_list
family1['utmk_y'] = y_list
```

```python
tttt = pd.merge(ttt, family1, on=['utmk_x', 'utmk_y'], how='left')
tttt
```

```python
tttt.rename(columns={tttt.columns[9]:"1인가구"},inplace=True)
tttt=tttt.drop('geometry',axis=1)
tttt
```

```python
tttt.dropna()
```

```python
data = tttt.copy()
```

```python
# Nan값 0대체
data =data.fillna(0)
```

```python
data
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
scaler.fit(data.loc[:,['거주인구','유동인구_평균','배달인구_평균','1인가구']])
```

```python
sc_data=pd.DataFrame(scaler.transform(data.loc[:,['거주인구','유동인구_평균','배달인구_평균','1인가구']]))
sc_data
```

```python
sc_data=sc_data.rename(columns={sc_data.columns[0] : 'sc_거주인구',sc_data.columns[1]  : 'sc_유동인구_평균',
                                sc_data.columns[2] : 'sc_배달인구_평균',sc_data.columns[3] : 'sc_1인가구'})
sc_data
```

```python
sc_data_m = pd.concat([data,sc_data],axis=1)
```

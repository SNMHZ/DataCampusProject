```python


file1 = '/content/drive/MyDrive/2019년_10살_배달인구.txt'
file2 = '/content/drive/MyDrive/2019년_20살_배달인구.txt'
file3 = '/content/drive/MyDrive/2019년_30살_배달인구.txt'
file4 = '/content/drive/MyDrive/2019년_40살_배달인구.txt'
file5 = '/content/drive/MyDrive/2019년_50살_배달인구.txt'
file6 = '/content/drive/MyDrive/2019년_60살_배달인구.txt'



col2_1 = []
col4_1 = []


f1 = open(file1, 'r')
line = f1.readline()

while True:

    line = f1.readline()
    if not line: 
      break

    l = line.split('^')

    col2_1.append(l[1])
    col4_1.append(float(l[3]))

#===================================================================


col2_2 = []
col4_2 = []


f2 = open(file2, 'r')
line = f2.readline()

while True:

    line = f2.readline()
    if not line: 
      break

    l = line.split('^')

    col2_2.append(l[1])
    col4_2.append(float(l[3]))

#===================================================================


col2_3 = []
col4_3 = []


f3 = open(file3, 'r')
line = f3.readline()

while True:

    line = f3.readline()
    if not line: 
      break

    l = line.split('^')

    col2_3.append(l[1])
    col4_3.append(float(l[3]))

#===================================================================


col2_4 = []
col4_4 = []


f4 = open(file4, 'r')
line = f4.readline()

while True:

    line = f4.readline()
    if not line: 
      break

    l = line.split('^')

    col2_4.append(l[1])
    col4_4.append(float(l[3]))

#===================================================================


col2_5 = []
col4_5 = []


f5 = open(file5, 'r')
line = f5.readline()

while True:

    line = f5.readline()
    if not line: 
      break

    l = line.split('^')
    col2_5.append(l[1])
    col4_5.append(float(l[3]))

#===================================================================

col2_6 = []
col4_6 = []


f6 = open(file6, 'r')
line = f6.readline()

while True:

    line = f6.readline()
    if not line: 
      break

    l = line.split('^')
    col2_6.append(l[1])
    col4_6.append(float(l[3]))
```

```python

import pandas as pd


df1 = pd.DataFrame({
    "집계구코드": col2_1,
    "10살배달인구": col4_1,
    })
df2 = pd.DataFrame({
    "집계구코드": col2_2,
    "20살배달인구": col4_2,
    })
df3 = pd.DataFrame({
    "집계구코드": col2_3,
    "30살배달인구": col4_3,
    })
df4 = pd.DataFrame({
    "집계구코드": col2_4,
    "40살배달인구": col4_4,
    })
df5 = pd.DataFrame({
    "집계구코드": col2_5,
    "50살배달인구": col4_5,
    })
df6 = pd.DataFrame({
    "집계구코드": col2_6,
    "60살배달인구": col4_6,
    })

df1.set_index('집계구코드', inplace = True)
df2.set_index('집계구코드', inplace = True)
df3.set_index('집계구코드', inplace = True)
df4.set_index('집계구코드', inplace = True)
df5.set_index('집계구코드', inplace = True)
df6.set_index('집계구코드', inplace = True)

print(df1)
print(df2)
print(df3)
print(df4)
print(df5)
print(df6)

x = df1.join(df2, how='right')
x = x.join(df3, how='right')
x = x.join(df4, how='right')
x = x.join(df5, how='right')
x = x.join(df6, how='right')
x = x.dropna(axis= 0)
x.index
```

```python
x.corr()
```

```python
from sklearn.preprocessing import MinMaxScaler

min_max_scaler = MinMaxScaler()
fitted = min_max_scaler.fit(x)
print(fitted.data_max_)
print(fitted.data_min_)

output = min_max_scaler.transform(x)
output = pd.DataFrame(output, columns=x.columns, index=list(x.index.values))
output = output.dropna(axis=0)
output
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(output)
PC_score = pca.transform(output)
PC_score


pc1 = []
pc2 = []
pc3 = []
pc4 = []
pc5 = []
pc6 = []


for i in range(len(PC_score)) :
  for j in range(6) :
    if j == 0 :
      pc1.append(PC_score[i][0])
    elif j == 1:
      pc2.append(PC_score[i][1])
    elif j == 2:
      pc3.append(PC_score[i][2])
    elif j == 3:
      pc4.append(PC_score[i][3])
    elif j == 4:
      pc5.append(PC_score[i][4])
    elif j == 5:
      pc6.append(PC_score[i][5])

pc_result = pd.DataFrame({
    'PC1': pc1,
    'PC2': pc2,
    'PC3': pc3,
    'PC4': pc4,
    'PC5': pc5,
    'PC6': pc6,

})
pc_result
```

```python
ratio = pca.explained_variance_ratio_
ratio


df_v = pd.DataFrame(ratio, index=['PC1','PC2','PC3','PC4', 'PC5', 'PC6'], columns=['V_ration'])
df_v.plot.pie(y='V_ration')
df_v
```

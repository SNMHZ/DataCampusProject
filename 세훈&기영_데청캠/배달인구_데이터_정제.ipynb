```python
col1 = []
col2 = []
col3 = []
col4 = []


file = "/content/drive/MyDrive/11_2019년_성연령별인구.txt"
f = open(file, 'r')
line = f.readline()

while True:

    line = f.readline()
    if not line: 
      break
    l = line.split('^')

    col1.append(l[0])
    col2.append(l[1])
    col3.append(l[2])
    col4.append(l[3])

f.close()
```

나이 정보
- 10대: in_age_003, in_age_004
- 20대: in_age_005, in_age_006
- 30대: in_age_007, in_age_008
- 40대: in_age_009, in_age_010
- 50대: in_age_011, in_age_012
- 60대: in_age_013, in_age_014
- 70대: in_age_015, in_age_016


```python
for i in range(len(col4)) :
  if col4[i] == 'N/A\n' or col4[i] == 'N/A' :
      col4[i] = '2.5'
  
  col4[i] = float(col4[i])
```

```python
import pandas as pd

df = pd.DataFrame({
    "집계구코드": col2,
    "나이코드": col3,
    "인원수": col4,
    })
```

```python

#10대
df_10 = df.loc[['in_age_003', 'in_age_004']]

#20대
df_20 = df.loc[['in_age_005', 'in_age_006']]
#30대
df_30 = df.loc[['in_age_007', 'in_age_008']]
#40대
df_40 = df.loc[['in_age_009', 'in_age_010']]
#50대
df_50 = df.loc[['in_age_003', 'in_age_012']]
#60대 이상
df_60up = df.loc[['in_age_013', 'in_age_014', 'in_age_015', 'in_age_016']]


for i in range(len(df_10)) :
  df_10.iloc[i,1] =  df_10.iloc[i,1] * 0.05


for i in range(len(df_20)) :
  df_20.iloc[i,1] =  df_20.iloc[i,1] * 0.29


for i in range(len(df_30)) :
  df_30.iloc[i,1] =  df_30.iloc[i,1] * 0.35

for i in range(len(df_40)) :
  df_40.iloc[i,1] =  df_40.iloc[i,1] * 0.22

for i in range(len(df_50)) :
  df_50.iloc[i,1] =  df_50.iloc[i,1] * 0.07

for i in range(len(df_60up)) :
  df_60up.iloc[i,1] =  df_60up.iloc[i,1] * 0.02
```

```python
col1 = []
col2 = []
col3 = []
col4 = []


file = "/content/drive/MyDrive/11_2019년_인구총괄(총인구).txt"
f = open(file, 'r')
line = f.readline()

while True:

    line = f.readline()
    if not line: 
      break
    l = line.split('^')
    l_2 = l[2].split('_')

    #10대와 70대 제외
    if int( int(l_2[2]) / 10) == 0 or int(int(l_2[2]) / 10) == 7 :
      continue

    col1.append(l[0])
    col2.append(l[1])
    col3.append(int(l_2[2]))
    col4.append(l[3])

f.close()
```
